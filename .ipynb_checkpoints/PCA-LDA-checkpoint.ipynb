# Biblioteca de Digitos

## Carregue a Base MNIST

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

### Normalizar dados

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_pca_std = scaler.fit_transform(X_train_pca.data)
X_test_pca_std = scaler.fit_transform(X_test_pca.data)
```

## PCA

### Aplique PCA avaliando o número de componentes necessários

```python
pca = PCA().fit(X_train_pca_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
pca = PCA(n_components=0.90)

X_pca = pca.fit_transform(X_train_pca_std)

principalDf = pd.DataFrame(data = X_pca)

principalDf.head()
```

```python
print(X_train_pca_std.shape)
print(X_pca.shape)
```

```python
print(X_pca)
```

### Visualize as imagens associadas com os primeiros componentes

```python
def plotImg(data):
    fig, axes = plt.subplots(6,39,figsize=(20,3),
    subplot_kw={'xticks':[], 'yticks':[]},
    gridspec_kw=dict(hspace=0.01, wspace=0.01))

    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(28,28))
```

```python
plotImg(data=X_train_pca)
```

```python
plotImg(data=X_train_pca_std)
```

```python
plotImg(pca.components_)
```

### Exiba a variância cumulativa

```python
pca.explained_variance_ratio_
```

```python
tot = sum(pca.explained_variance_ratio_)
var_exp = [(i / tot)*100 for i in sorted(pca.explained_variance_ratio_, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
cum_var_exp
```

```python
plt.plot(cum_var_exp)
plt.xlabel('Número de Componentes')
plt.ylabel('Variância cumulativa')
```

### Compare as imagens de entrada com as imagens reconstruídas

```python
# Imagem Original
plotImg(data=X_train_pca)
```

```python
# Imagem Original Normalizada
plotImg(data=X_train_pca_std)
```

```python
# Imagem depois de passar pelo pca
plotImg(data=pca.components_)
```

```python
projection = pca.inverse_transform(X_pca)

plotImg(data=projection)
```

## LDA

### Aplique LDA nos dados originais

```python
X_train_lda, X_test_lda, y_train_lda, y_test_lda = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()

X_train_lda_std = scaler.fit_transform(X_train_lda.data)
X_test_lda_std = scaler.fit_transform(X_test_lda.data)
```

```python
lda = LDA(n_components=9)
X_lda = lda.fit_transform(X_train_lda_std, y_train_lda)
```

```python
X_lda
```

```python
X_lda.shape
```

```python
y_pred = lda.predict(X_test_lda_std)

y_pred
```

### Exiba a projeção dos dados nos Discriminantes Lineares

```python
le = LabelEncoder()
y_plot = le.fit_transform(y_train_lda)

plt.xlabel('Discriminante Linear 1')
plt.ylabel('Discriminante Linear 2')

plt.scatter(
    X_lda[:,0],
    X_lda[:,1],
    c= y_plot,
    cmap='rainbow',
    alpha=0.6,
    edgecolors='b'
)
```

### Exiba a taxa de variância acumulada

```python
tot = sum(lda.explained_variance_ratio_)
var_exp = [(i / tot)*100 for i in sorted(lda.explained_variance_ratio_, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
cum_var_exp
```

```python
plt.plot(cum_var_exp)
plt.ylabel('Variância cumulativa')
plt.xlabel('Componentes')
```

## Classificador

### Aplique um classificador de sua escolha nas features obtidas por LDA e PCA

```python
rfc = RandomForestClassifier(random_state=42)

param_grid = { 
    'n_estimators': [100, 200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

### GridSearchCv para Escolher os padrões que serão utilizados no nosso RFC

```python
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5, verbose=10)

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.3, random_state=42)

CV_rfc.fit(X_train_pca, y_train_pca)

print(CV_rfc.best_params_)
```

```python
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5, verbose=10, n_jobs=-3)

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_pca_std = scaler.fit_transform(X_train_pca.data)
X_test_pca_std = scaler.fit_transform(X_test_pca.data)

pca = PCA(n_components=0.90)
X_pca = pca.fit_transform(X_train_pca_std)

CV_rfc.fit(X_pca, y_train_pca)

print(CV_rfc.best_params_)
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_pca_std = scaler.fit_transform(X_train_pca.data)
X_test_pca_std = scaler.fit_transform(X_test_pca.data)

pca = PCA(n_components=0.90)
X_pca = pca.fit_transform(X_train_pca_std)

rfc_pca = RandomForestClassifier(criterion='entropy',max_depth=8,max_features='auto',n_estimators=500,
                                 random_state=42)

rfc_pca.fit(X_pca, y_train_pca)

X_test_pca_after = pca.transform(X_test_pca_std)
```

```python
y_pred = rfc_pca.predict(X_test_pca_after)

cfm = confusion_matrix(y_test_pca, y_pred)

report = classification_report(y_test_pca, y_pred)
print(report)

sns.heatmap(cfm, cbar=False, annot=True, cmap="Blues", fmt="d")
```

```python
X_train_lda, X_test_lda, y_train_lda, y_test_lda = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_lda_std = scaler.fit_transform(X_train_lda.data)
X_test_lda_std = scaler.fit_transform(X_test_lda.data)

lda = LDA(n_components=9)
X_lda = lda.fit_transform(X_train_lda_std, y_train_lda)

rfc_lda = RandomForestClassifier(criterion='entropy',max_depth=8,max_features='auto',n_estimators=500,
                                 random_state=42)

rfc_lda.fit(X_lda, y_train_lda)

X_test_lda_after = lda.transform(X_test_lda_std)
```

```python
y_pred = rfc_lda.predict(X_test_lda_after)

cfm = confusion_matrix(y_test_lda, y_pred)

report = classification_report(y_test_lda, y_pred)
print(report)

sns.heatmap(cfm, cbar=False, annot=True, cmap="Blues", fmt="d")
```
